```python
import modvis.ats_xdmf as xdmf
import modvis.plot_vis_file as pv
import modvis

model_dir = "../data/model_output"
```

```python
import os
if not os.path.exists(model_dir):
  !git clone https://github.com/pinshuai/modvis.git
  %cd ./modvis/examples/notebooks
```

# import vis data

```python
visfile = xdmf.VisFile(model_dir, domain='surface', load_mesh=True, model_time_unit='d')
```

# plot surface ponded depth

```python
# change the time_slice to obtain the surface ponded depth of all the plots used in the report

fig, ax, tpc = pv.plot_surface_data(visfile, var_name="surface-ponded_depth", 
                                    log = True, vmin=0.01, vmax=4, 
                                    time_slice= "2016-01-01")
```

```python
fig.patch.set_facecolor('white')
ax.set_facecolor('white')
fig.savefig("../figures/surface_ponded_depth_plot.png", dpi=500)
```

# plot ET

```python
# change the time_slice to obtain the surface total evapotranspiration of all the plots used in the report

fig, ax, tpc = pv.plot_surface_data(visfile, var_name="surface-total_evapotranspiration", 
                                    log = False, vmin=0, vmax=3, 
                                    time_slice= "2015-10-01")
```
